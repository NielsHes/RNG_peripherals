```python
import numpy as np
from pynput import mouse
from pynput.mouse import Button, Controller
from pynput.keyboard import Listener
```

```python
mouse = Controller()

# Read pointer position
print('The current pointer position is {}'.format(
    mouse.position))

# Set pointer position
mouse.position = (10, 20)
print('Now we have moved it to {}'.format(
    mouse.position))

# Move pointer relative to current position
mouse.move(5, -5)

# Press and release
mouse.press(Button.left)
mouse.release(Button.left)

# Double click; this is different from pressing and releasing
# twice on macOS
mouse.click(Button.left, 2)

# Scroll two steps down
mouse.scroll(0, 2)
```

```python
def on_press(key):
    print(f'Key {key} pressed')  # Print pressed key

def on_release(key):
    print(f'Key {key} released')  # Print released key
    if str(key) == 'Key.esc':     # Stop listener on ESC
        return False

# Start the listener
with Listener(on_press=on_press, on_release=on_release) as listener:
    listener.join()
```
